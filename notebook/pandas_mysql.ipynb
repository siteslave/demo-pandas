```python
import sqlalchemy
import pandas as pd

conn = 'mysql://root:789124@localhost:3306/cmi'
engine = sqlalchemy.create_engine(conn)
```

```python
hospcode = ("11053", "11008", "10670")

query = open("sql/01_select_cmi.sql", "r")
with engine.connect().execution_options(autocommit=True) as conn:
  df = pd.read_sql(sql=query.read(), con=engine, params={"hospcode": hospcode})

df
```

```python
from datetime import datetime

now = datetime.now()
str_time = now.strftime("%Y%m%d%H%M%S")
csv_filename = f"cmi_{str_time}.csv"

df.to_csv(csv_filename, sep="|", index=False)
```
