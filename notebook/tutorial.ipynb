```python
import pandas as pd

url = "https://raw.githubusercontent.com/paskhaver/pandas-in-action/master/chapter_06_working_with_text_data/customers.csv"

customers = pd.read_csv(url)
```

```python
customers["Name"].str.split(
 pat = " ", n = 1, expand = True
 ).head()
```

```python
customers[["First Name", "Last Name"]] = customers[
 "Name"
].str.split(pat = " ", n = 1, expand = True)

customers.head()
```

```python
customers = customers.drop(labels = "Name", axis = "columns")
```

```python
customers.head()
```

```python
customers["Address"].str.split(", ", expand = True).head()
```

```python
new_cols = ["Street", "City", "State", "Zip"]

customers[new_cols] = customers["Address"].str.split(
 pat = ", ", expand = True
)
```

```python
customers.head()
```

```python
customers.drop(labels = "Address", axis = "columns").head()
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/paskhaver/pandas-in-action/master/chapter_08_reshaping_and_pivoting/sales_by_employee.csv"

sales = pd.read_csv(url, parse_dates=["Date"])
```

```python
sales.head()
```

```python
import numpy as np
```

```python
sales.pivot_table(
  index="Date",
  values=["Revenue", "Expenses"], 
  aggfunc={"Expenses": np.mean, "Revenue": [np.mean, np.min, np.max]})
```

```python
sales.pivot_table(
  index = "Date",
  columns = "Name",
  values = "Revenue",
  fill_value=0,
  aggfunc = "sum"
 )
```

```python
sales.pivot_table(
 index = "Date",
 columns = "Name",
 values = "Revenue",
 aggfunc = "sum",
 fill_value = 0,
 margins = True
 )
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/paskhaver/pandas-in-action/master/chapter_09_the_groupby_object/fortune1000.csv"

fortune = pd.read_csv(url)
```

```python
fortune.head()
```

```python
sectors = fortune.groupby("Sector")
```

```python
sectors.size()
```

```python
sectors.first().head(10)
```

```python
sectors.sum(numeric_only=True).head(10)
```

```python
sectors["Revenues"].sum(numeric_only=True).head(10)
```

```python
aggregations = {
 "Revenues": "min",
 "Profits": "max",
 "Employees": "mean"
 }

sectors.agg(aggregations).head()
```

```python
fortune.nlargest(n = 5, columns = "Profits")
```
