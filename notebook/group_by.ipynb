```python
import pandas as pd
import numpy as np

groups1 = pd.read_csv("https://raw.githubusercontent.com/siteslave/pandas-in-action/master/chapter_10_merging_joining_and_concatenating/meetup/groups1.csv")

groups2 = pd.read_csv("https://raw.githubusercontent.com/siteslave/pandas-in-action/master/chapter_10_merging_joining_and_concatenating/meetup/groups2.csv")

categories = pd.read_csv("https://raw.githubusercontent.com/siteslave/pandas-in-action/master/chapter_10_merging_joining_and_concatenating/meetup/categories.csv")

cities = pd.read_csv("https://raw.githubusercontent.com/siteslave/pandas-in-action/master/chapter_10_merging_joining_and_concatenating/meetup/cities.csv", dtype = {"zip": "string"})
```
